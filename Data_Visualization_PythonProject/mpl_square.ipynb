Plotting a Simple Line Graph

```python
import matplotlib.pyplot as plt
squares = [1, 4, 9, 16, 25]
plt.plot(squares)
plt.show()
```

_One of the simplest plots you can make in matplotlib_

Changing the Label Type and Graph Thickness

```python
plt.plot(squares, linewidth=5)
# Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
# Set size of tick labels.
plt.tick_params(axis='both', labelsize=14)
plt.show()
```

_The chart is much easier to read now_

Correcting the Plot

Notice at the end of the graph that the square of 4.0 is shown 
as 25! Lets fix that

```python
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.plot(input_values, squares, linewidth=5)
plt.show()
```

_The data is now plotted correctly_
